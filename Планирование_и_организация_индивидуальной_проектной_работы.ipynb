**Часть I. Инструкция по постановке целей и задач проекта.**

Цель: разработать объектно-ориентированную версию шахматного симулятора, которая реализует все правила перемещения фигур и их взаимодействие, создать объекты и абстрактные сущности игры в шахматы.

Задачи:
1. Представить шахматные фигуры, доску, ходы фигур в виде объектов;
2. Описать атрибуты классов каждой фигуры, взаимодействия фигур, ходов по шахматной доске, случаи выигрыша и поражения сторон;
3. Дополнить основные действия игры дополнительными заданиями;
4. Проверить работу классов и их атрибутов и методов, исправить ошибки.
